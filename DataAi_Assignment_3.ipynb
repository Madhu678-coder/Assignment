```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
```

```python
df_processed = df.drop("User ID", axis=1)
df_processed["Gender"] = LabelEncoder().fit_transform(df_processed["Gender"])
```

```python
df_processed.corr()
```

```python
X = df_processed.drop("Purchased", axis=1)
y = df_processed["Purchased"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
scaler_minmax = MinMaxScaler()
X_train_minmax = scaler_minmax.fit_transform(X_train)
X_test_minmax = scaler_minmax.transform(X_test)
```

```python
scaler_standard = StandardScaler()
X_train_standard = scaler_standard.fit_transform(X_train)
X_test_standard = scaler_standard.transform(X_test)
```

```python
clf_minmax = DecisionTreeClassifier(random_state=0)
clf_minmax.fit(X_train_minmax, y_train)
y_pred_minmax = clf_minmax.predict(X_test_minmax)
acc_minmax = accuracy_score(y_test, y_pred_minmax)
print(f"Accuracy with MinMaxScaler:, {acc_minmax:.4f}")
```

```python
clf_standard = DecisionTreeClassifier(random_state=0)
clf_standard.fit(X_train_standard, y_train)
y_pred_standard = clf_standard.predict(X_test_standard)
acc_standard = accuracy_score(y_test, y_pred_standard)
print(f"Accuracy with StandardScaler:, {acc_standard:.4f}")
```
